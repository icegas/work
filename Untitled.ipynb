```python
import pandas as pd
import numpy as np
import seaborn as sn
import pywt as wt
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import (RBF, Matern, RationalQuadratic,
                                              ExpSineSquared, DotProduct,
                                              ConstantKernel)
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingRegressor
```

```python
X = np.random.uniform(size=(80, 1000))
x = np.array([np.sum(X[i,:]) for i in range(X.shape[0])])
Y = np.sin(x*6+np.cos(2*x)) * np.sin(2*x) + np.random.normal(0, 0.1,size=80)
```

```python
def get_f(df):
    m = df.mean()
    s = df.std()
    p5 = df.quantile(0.05)
    p95 = df.quantile(0.95)
    
    df_wav = df.apply(wt.wavedec, args=('db4', 'sym', 5))
    
    data = np.array([])
    for wav in df_wav:
        for c in wav:
            data = np.append(data, [c.mean(), c.std()])

    data = pd.DataFrame( np.reshape(data, (80, -1)) )
            
    
    return pd.concat([m, s, p5, p95, data], axis=1)
```

```python
df_new = get_f(df)
df_y = pd.DataFrame(Y)
f_df = pd.concat([df_new, df_y], axis=1)
```
